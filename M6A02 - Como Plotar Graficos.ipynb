# Gráfico de Linhas

* Normalmente utilizado para visualização de dados em uma linha do tempo.
* Normalmente: eixo X - Dados temporais eixo Y - dados quantitatívos.

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

**Definindo label aos eixos**

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.show()
```

**Alterando a forma dos pontos**

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'o')
plt.show()
```

**Definindo limites de visualização dos eixos x e y**

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'o')
plt.axis([0, 6, 0, 20])
plt.show()
```

**Plot duas séries de dados, no caso x1 e y2 e x3 e y3**
